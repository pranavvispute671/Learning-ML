```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv("../Polynomial Linear Regression/Position_Salaries.csv")
```

```python
x = data.iloc[:,1:-1].values
y = data.iloc[:,-1].values
```

```python
regressor = RandomForestRegressor(n_estimators=12, random_state=0)
```

```python
regressor.fit(x, y)
```

```python
print(regressor.predict([[6]]))
```

```python
Xgrid = np.arange(min(x), max(x), 0.1)
Xgrid = Xgrid.reshape((len(Xgrid),1))
plt.scatter(x,y, color ="red")
plt.plot(Xgrid, regressor.predict(Xgrid), color="blue")
plt.title('Decision tree Regression')
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```
